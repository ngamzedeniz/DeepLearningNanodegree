```python
!wget https://ocw.mit.edu/ans7870/6/6.006/s08/lecturenotes/files/t8.shakespeare.txt

with open("t8.shakespeare.txt","r") as sh_fd:
    sheakspeare = " ".join(sh_fd.readlines())

find_this_string = sheakspeare[:100].lower()

from random import choice
from string import ascii_lowercase

# vary the number of spaces appended to adjust the probability
chars = ascii_lowercase + " " * 10

def random_string(n):
    return "".join(choice( list(set(find_this_string)) ) for _ in range(n))


def fitness(truth, prediction):
    counter = 0
    for letter_idx in range(len(truth)):
        if truth[letter_idx] != prediction[letter_idx]:
            counter += 1
    return counter

#def mutate(gene, mutation_frequency = 0.01):
#    return 
from random import random, randint
from copy import copy
def tournament_crossover(c1, c2, r_cross=0.5, mutation_p = 0.1):
	# children are copies of parents by default
	# check for recombination
    new_child = ""
    for c_idx in range(len(c1)):
        if random() <= mutation_p:
            new_child += random_string(1)
        elif random() < r_cross:
            #print("mother")
            new_child += c1[c_idx]
        else:
            #print("father")
            new_child += c2[c_idx]

    return new_child

def tournament_mutate(parent, mutation_count = 1):
    new_parent = ""
    for _ in range(mutation_count):#FIXME
        idx_to_mutate =randint(0,len(parent))
        for idx in range(len(parent)):
            if idx == idx_to_mutate:
                new_parent += random_string(1)
            else:
                new_parent += parent[idx]
    return new_parent

from random import choice
from copy import deepcopy
def repopulate(top_performers,possibility_of_mutation =0.01, target_population_size = 1000):
    new_population = deepcopy(top_performers)
    while len(new_population) < target_population_size:
        p1 = choice(top_performers)
        p2 = choice(top_performers)
        child = crossover(p1,p2,0.5)
        new_population.append(child)
    return new_population

import numpy as np
def evolve(inital_seeder = random_string, fitness, new_percentage = 0.8, population=100):
    
    ### initialize population
    gene_pool = np.array([inital_seeder(len(find_this_string)) for _ in range(population)])
    
    for _ in range(100000):
        fitness_results = np.array([fitness(find_this_string,gene) for gene in gene_pool])
        
        index_of_top = fitness_results.argmin()
        index_of_bottom = fitness_results.argmax()
        random_mate = randint(0, len(gene_pool)-1)
        child = tournament_crossover(gene_pool[index_of_top],gene_pool[random_mate] )

        ### drop worse
        gene_pool = np.delete(gene_pool, index_of_bottom)
        ### mutate best performer
        if random() < new_percentage:
            gene_pool = np.insert(gene_pool, 0, child)
        else: 
            gene_pool = np.insert(gene_pool, 0, inital_seeder(len(find_this_string)))

        if _ % 100 == 0:
            print(gene_pool[index_of_top])
            print("Best performer: {}".format(fitness_results[index_of_top]))

        
```
